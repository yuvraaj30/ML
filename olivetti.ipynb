# plot the digits: image is 8x8 grid and label the image with the target value

```python
# Display sample images
sample_indices = [0, 40, 80, 120]  # Select images from four classes (every 40th)
num_rows, num_cols = 2, 2  # Arrange in a 2x2 grid

for i in range(num_rows):
    for j in range(num_cols):
        fig, ax = plt.subplots()
        img = df.data[sample_indices[i * num_cols + j]].reshape((64, 64))  # Reshape pixel values
        ax.imshow(img, cmap='gray')
        ax.set_title(f"Class {targets[sample_indices[i * num_cols + j]]}")
        ax.axis('off')
        plt.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust spacing for better visualization

plt.show()
```

```python
# Plot the digits: image is 8x8 grid and label the image with the target value
plt.figure(figsize=(10, 4))
for i in range(10):  # Display first 10 images
    plt.subplot(2, 5, i + 1)
    plt.imshow(features[i].reshape(64, 64), cmap=plt.cm.gray)
    plt.title(f"Target: {targets[i]}")
    plt.axis('off')
```
