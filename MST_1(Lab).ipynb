Build a Basic python web application to check given number is even or odd.(please guys run this code on VS code)

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/")
def index():
    number = request.args.get('number', "")
    result = ""

    if number:
        try:
            result = even_or_odd(int(number))
        except ValueError:
            result = "Invalid input. Please enter a valid number."

    return (
        '''<form action="" method="get">
                Enter a number: <input type="text" name="number">
                <input type="submit" value="Check Odd/Even">
            </form>'''
        + "Result: "
        + result
    )

def even_or_odd(number):
    return "Even" if number % 2 == 0 else "Odd"

if __name__=="__main__":
    app.run(host="127.0.0.1", port=9090, debug=True)
    
```

# Lab 1.1_2
Build a Basic python web application to display given number is prime or not. 

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/")
def index():
    number = request.args.get('number', "")
    result = ""

    if number:
        try:
            result = is_prime(int(number))
        except ValueError:
            result = "Invalid input. Please enter a valid number."

    return (
        '''<form action="" method="get">
                Enter a number: <input type="text" name="number">
                <input type="submit" value="Check Prime">
            </form>'''
        + "Result: "
        + result
    )

def is_prime(number):
    """Check if the number is prime."""
    if number <= 1:
        return "Not Prime"
    if number <= 3:
        return "Prime"
    if number % 2 == 0 or number % 3 == 0:
        return "Not Prime"
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return "Not Prime"
        i += 6
    return "Prime"

if __name__=="__main__":
    app.run(host="127.0.0.1", port=9090, debug=True)
```

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/", methods=['GET', 'POST'])
def index():
    result = ""
    if request.method == 'POST':
        try:
            year = int(request.form['year'])
            result = is_leap_year(year)
        except ValueError:
            result = "Invalid input. Please enter a valid year."

    return (
        '''<form action="" method="post">
                Enter a year: <input type="text" name="year">
                <input type="submit" value="Check Leap Year">
            </form>'''
        + "Result: "
        + result
    )

def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return "Leap Year"
            else:
                return "Not a Leap Year"
        else:
            return "Leap Year"
    else:
        return "Not a Leap Year"

if __name__=="__main__":
    app.run(host="127.0.0.1", port=9090, debug=True)
```

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/", methods=['GET', 'POST'])
def index():
    result = ""
    if request.method == 'POST':
        try:
            number = int(request.form['number'])
            result = is_divisible_by_6(number)
        except ValueError:
            result = "Invalid input. Please enter a valid number."

    return (
        '''<form action="" method="post">
                Enter a number: <input type="text" name="number">
                <input type="submit" value="Check Divisibility by 6">
            </form>'''
        + "Result: "
        + result
    )

def is_divisible_by_6(number):
    if number % 6 == 0:
        return "Divisible by 6"
    else:
        return "Not divisible by 6"

if __name__=="__main__":
    app.run(host="127.0.0.1", port=9010, debug=True)
```
